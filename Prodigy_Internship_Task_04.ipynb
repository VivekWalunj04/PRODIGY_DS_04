```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col_names = ["ID","Entity","Sentiments","Contest"]
data = pd.read_csv('/content/twitter_training.csv',names=col_names)
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data.dropna(axis=0,inplace=True)
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
data.shape
```

```python
sentiment_counts = data["Sentiments"].value_counts()
sentiment_counts
```

```python
plt.figure(figsize=(6,3))
sentiment_counts.plot(kind='bar',color=["red","green","blue","yellow"])
plt.title("sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Tweets")
```

```python
brand_data = data[data["Entity"].str.contains("Google",case=False)]
brand_sentiment_counts = brand_data["Sentiments"].value_counts()
brand_sentiment_counts
```

```python
plt.figure(figsize=(6,6))
plt.pie(brand_sentiment_counts,labels=brand_sentiment_counts.index,autopct='%1.1f%%',colors=["red","green","blue","yellow"])
plt.show()
```
